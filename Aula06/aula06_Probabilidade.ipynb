```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# Lendo os arquivos de dados com a biblioteca pandas 
df_2020 = pd.read_excel('/content/respostas_embaixadinha_2020.xlsx')
df_2021 = pd.read_excel('/content/respostas_embaixadinha_2021.xlsx')
display(df_2020.head(2))
display(df_2021.head(2))
```

```python
# Removendo a coluna de Endereço de e-mail, pois ela está sem informações
df_2020.drop(columns=['Endereço de e-mail'], inplace=True)
#df_2020.drop(columns=['Endereço de e-mail','Carimbo de data/hora'], inplace=True)

df_2021.drop(columns=['Endereço de e-mail'], inplace=True)
#df_2021.drop(columns=['Endereço de e-mail','Carimbo de data/hora'], inplace=True)

#display(df_2020.head(2))
#display(df_2021.head(2))
```

```python
# Removento valores faltantes NaN
print(f'Antes: \n2020 = {df_2020.shape}\n2021 = {df_2021.shape}')
df_2020.dropna(subset=['Qual seu número USP?'], inplace=True)
df_2021.dropna(subset=['Qual seu número USP?'], inplace=True)
print(f'Depois: \n2020 = {df_2020.shape}\n2021 = {df_2021.shape}')
```

```python
# Converter column 'Qual seu número USP?' para inteiro
df_2020['Qual seu número USP?'] = df_2020['Qual seu número USP?'].astype(int)
df_2021['Qual seu número USP?'] = df_2021['Qual seu número USP?'].astype(int)
```

```python
# Removendo quem tem número USP duplicado, mantendo o último e apagando os anteriores
df_2020.drop_duplicates(subset=['Qual seu número USP?'], inplace=True, keep='last')
df_2021.drop_duplicates(subset=['Qual seu número USP?'], inplace=True, keep='last')
```

# Probabilidade


1.   Qual a probabilidade de escolher uma pessoa do sexo F de cada sala?
2.   Qual sala eu tenho mais chances de escolher alguém de skill de 5?

```python
total_2020 = df_2020.shape[0]
total_2021 = df_2021.shape[0]
print(f'O total de alunos em 2020 eh {total_2020}\nO total de alunos em 2021 eh {total_2021}')
```

**Resposta: 1. Qual a probabilidade de escolher uma pessoa do sexo F de cada sala?**



```python
# Contar quantos são do sexo masculino (M) e quantos são do sexo feminino (F)
mf_2020 = df_2020['Sexo (M = masculino ou F = feminino)'].value_counts()
mf_2021 = df_2021['Sexo (M = masculino ou F = feminino)'].value_counts()
print(f'2020:\n{mf_2020}\n2021:\n{mf_2021}')
```

```python
# A prob de escolher uma mulher em 2020 eh...
prob_2020 = mf_2020[1] / total_2020
prob_2020 = round(prob_2020, 2) # para arredondar com duas casas decimais
print(f'A probabilidade de escolher ao acaso uma mulher da turm@ de 2020 eh de:\n{100*prob_2020}%')
```

```python
# A prob de escolher uma mulher em 2021 eh...
prob_2021 = mf_2021[1] / total_2021
prob_2021 = round(prob_2021, 2) # para arredondar com duas casas decimais
print(f'A probabilidade de escolher ao acaso uma mulher da turm@ de 2021 eh de:\n{100*prob_2021}%')
```
