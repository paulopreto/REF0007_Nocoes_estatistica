# Probabilidade


1.   Qual a probabilidade de escolher uma pessoa do sexo F de cada sala?
2.   Qual sala eu tenho mais chances de escolher alguém de skill de 5?

```python
total_2020 = df_2020.shape[0]
total_2021 = df_2021.shape[0]
print(f'O total de alunos em 2020 eh {total_2020}\nO total de alunos em 2021 eh {total_2021}')
```

**Resposta: 1. Qual a probabilidade de escolher uma pessoa do sexo F de cada sala?**



```python
# Contar quantos são do sexo masculino (M) e quantos são do sexo feminino (F)
mf_2020 = df_2020['Sexo (M = masculino ou F = feminino)'].value_counts()
mf_2021 = df_2021['Sexo (M = masculino ou F = feminino)'].value_counts()
print(f'2020:\n{mf_2020}\n2021:\n{mf_2021}')
```

```python
# A prob de escolher uma mulher em 2020 eh...
prob_2020 = mf_2020[1] / total_2020
prob_2020 = round(prob_2020, 2) # para arredondar com duas casas decimais
print(f'A probabilidade de escolher ao acaso uma mulher da turm@ de 2020 eh de:\n{100*prob_2020}%')
```

```python
# A prob de escolher uma mulher em 2021 eh...
prob_2021 = mf_2021[1] / total_2021
prob_2021 = round(prob_2021, 2) # para arredondar com duas casas decimais
print(f'A probabilidade de escolher ao acaso uma mulher da turm@ de 2021 eh de:\n{100*prob_2021}%')
```

**Resposta: 2.Qual sala eu tenho mais chances de escolher alguém de nível de habilidade (skill) de 5?**

```python
# Contar quantos são dos níveis de habilidade de 1 - 5
skill_2020 = df_2020['Em uma escala entre 1 e 5, qual o seu nível de habilidade no futebol?'].value_counts()
skill_2021 = df_2021['Em uma escala entre 1 e 5, qual o seu nível de habilidade no futebol?'].value_counts()
```

```python
print(f'Em 2020 o skill foi:\n{skill_2020}\n\nEm 2021 o skill foi:\n{skill_2021}')
```

```python
# df_2020.dropna(subset=['Em uma escala entre 1 e 5, qual o seu nível de habilidade no futebol?'], inplace=True)
totalskill_2020 = df_2020['Em uma escala entre 1 e 5, qual o seu nível de habilidade no futebol?'].shape[0]
totalskill_2021 = df_2021['Em uma escala entre 1 e 5, qual o seu nível de habilidade no futebol?'].shape[0]
print(f'Total\n2020 = {totalskill_2020}\n2021 = {totalskill_2021}')
```

```python
#                     favorável   /   total             
prob_skill1_2020 = skill_2020[1] / totalskill_2020
prob_skill1_2021 = skill_2021[1] / totalskill_2021

prob_skill1_2020_porce = round(prob_skill1_2020, 2) * 100
prob_skill1_2021_porce = round(prob_skill1_2021, 2) * 100

print(f'A probabilidade de um skill 1 em:\n2020 = {prob_skill1_2020}\n2021 = {prob_skill1_2021}')
print(f'\nA probabilidade de um skill 1 em %:\n2020 = {prob_skill1_2020_porce}%\n2021 = {prob_skill1_2021_porce}%')
```

```python
#                     favorável   /   total             
prob_skill2_2020 = skill_2020[2] / totalskill_2020
prob_skill2_2021 = skill_2021[2] / totalskill_2021
		  
prob_skill2_2020_porce = round(prob_skill2_2020, 2) * 100
prob_skill2_2021_porce = round(prob_skill2_2021, 2) * 100

print(f'A probabilidade de um skill 2 em:\n2020 = {prob_skill2_2020}\n2021 = {prob_skill2_2021}')
print(f'\nA probabilidade de um skill 2 em %:\n2020 = {prob_skill2_2020_porce}%\n2021 = {prob_skill2_2021_porce}%')
```
