# Probabilidade


1.   Qual a probabilidade de escolher uma pessoa do sexo F de cada sala?
2.   Qual sala eu tenho mais chances de escolher alguém de skill de 5?

```python
total_2020 = df_2020.shape[0]
total_2021 = df_2021.shape[0]
print(f'O total de alunos em 2020 eh {total_2020}\nO total de alunos em 2021 eh {total_2021}')
```

**Resposta: 1. Qual a probabilidade de escolher uma pessoa do sexo F de cada sala?**



```python
# Contar quantos são do sexo masculino (M) e quantos são do sexo feminino (F)
mf_2020 = df_2020['Sexo (M = masculino ou F = feminino)'].value_counts()
mf_2021 = df_2021['Sexo (M = masculino ou F = feminino)'].value_counts()
print(f'2020:\n{mf_2020}\n2021:\n{mf_2021}')
```

```python
# A prob de escolher uma mulher em 2020 eh...
prob_2020 = mf_2020[1] / total_2020
prob_2020 = round(prob_2020, 2) # para arredondar com duas casas decimais
print(f'A probabilidade de escolher ao acaso uma mulher da turm@ de 2020 eh de:\n{100*prob_2020}%')
```

```python
# A prob de escolher uma mulher em 2021 eh...
prob_2021 = mf_2021[1] / total_2021
prob_2021 = round(prob_2021, 2) # para arredondar com duas casas decimais
print(f'A probabilidade de escolher ao acaso uma mulher da turm@ de 2021 eh de:\n{100*prob_2021}%')
```

**Resposta: 2.Qual sala eu tenho mais chances de escolher alguém de nível de habilidade (skill) de 5?**

# Pergunta Tarefa 3: Demonstre qual é a probabilidade (df_2020 e df_2021) que temos de escolher uma aluna (sexo), com menos de 20 anos (idade), com menos de 1.70 m (estatura) e que tenha feito mais de 5 embaixadinhas? 
## É possível descobrir esta informação?

```python
total_2020 = len(df_2020['Sexo (M = masculino ou F = feminino)'])
alunas_2020 = df_2020['Sexo (M = masculino ou F = feminino)'].value_counts().F / total_2020
print(f'Probabilidade de F = {alunas_2020}')

total_2021 = len(df_2021['Sexo (M = masculino ou F = feminino)'])
alunas_2021 = df_2021['Sexo (M = masculino ou F = feminino)'].value_counts().F / total_2021
print(f'Probabilidade de F = {alunas_2021}')
```

```python
# df_2020 = pd.read_excel('/content/respostas_embaixadinha_2020.xlsx', converters={'Qual sua data de nascimento?':pd.to_datetime})
# df_2020['nascimento'] = pd.to_datetime(df_2020['Qual sua data de nascimento?'])
# df_2021['nascimento'] = pd.to_datetime(df_2021['Qual sua data de nascimento?']).dt.date
df_2020['nascimento'] = pd.to_datetime(df_2020['Qual sua data de nascimento?'], format='%Y-%m-%d')
#display(df_2020.head(2))
today = pd.to_datetime('today')

tempo_vida = (today - df_2020['nascimento'])
#print(tempo_vida.dt.days)
df_2020['idade'] = tempo_vida.dt.days / 365.25

# display(df_2020.head(5))

menorde20_2020 = (df_2020['idade'] < 20).sum()
#bool_menorde20_2020 = menorde20_2020.sum()


prob_menorde20 = menorde20_2020 / total_2020
print(f'Probabilidade de menos de 20 anos = {prob_menorde20}')
```

```python
probm170_2020 = (df_2020['Qual sua estatura em metros?'] < 1.70).sum() / total_2020
#print((df_2020['Qual sua estatura em metros?'] < 1.70).value_counts()[1])
print(f'Probabilidade de menor de 1.70 m = {probm170_2020}')
```

```python
probm5emba_2020 = (df_2020['Quantas embaixadinhas você conseguiu fazer?'] > 5).sum() / total_2020
#print((df_2020['Qual sua estatura em metros?'] > 5).value_counts()[1])
print(f'Probabilidade de mais de 5 embaixadinhas = {probm5emba_2020}')
```
